```python
%%writefile pipelines.py
from pprint import pprint
import csv
from scrapy.pipelines.images import ImagesPipeline
import scrapy

#В этот раз с пепилинами.
# В этом пипелине обрабатываем ссылки - по инструкции из семинара
class SimplePhotosPipeline(ImagesPipeline):
    def get_media_requests(self,item,info):
        if item['photos']:
            for img_url in item['photos']:
                try:
                    yield scrapy.Request(img_url)#Наполучали фоток!
                except Exception as e:
                    print(e)
                    
    def item_completed(self,results,item,info):
        # pprint(result)
        if results:
            item['photos']=[itm[1] for itm in results if itm[0]] #results содержит всякое про results
        return item
        
#В этом пипелине записываем наработаное в csv-файл
class SimplePipeline:

    def open_spider(self, spider):
        self.file = open('output.csv', 'w', newline='')
        self.writer = csv.DictWriter(self.file, fieldnames=['title', 'author', 'url','published','photos'])
        self.writer.writeheader() #Запись заголовков
        pass

    def process_item(self, item, spider):
        # Записываем данные в CSV файл
        self.writer.writerow(item) #Запись данных
        #print(item)  # Выводим данные на экран - контролировали процесс - держали руку на пульсе
        return item

    def close_spider(self, spider):
        self.file.close()  # Закрываем файл после завершения работы паука
        print("Done")
```

```python
%%writefile myspider.py
import scrapy
from scrapy.http import HtmlResponse
from pymongo.mongo_client import MongoClient
from bookitem import SimpleItem
from scrapy.loader import ItemLoader


class UnsplashSpider(scrapy.Spider):
    name = 'unsplash'
    allowed_domains=["unsplash.com"]
    
    def __init__(self,**kwargs):
        super().__init__(**kwargs)
        self.start_urls = [f"https://unsplash.com/s/photos/{kwargs.get('category')}"] # Можно указать категорию фоток
        
    #Парсим данные по инструкции
    def parse(self, response:HtmlResponse):
        # print(response.status,response.url)
        # links=response.xpath("//a[@class='zNNw1']")
        links=response.xpath("//a[@class='zNNw1']/@href")
        for link in links:
            yield response.follow(link,callback=self.parse_book) #может сам доставать ссылку из элементов (работает не всегда) - в этот раз сработал!
        
                       
    def parse_book(self,response:HtmlResponse):
        #Этот и другой закоменченый код использовал при ведении настроечных работ!
        # item{
        #     'title':response.xpath("//p[@class='liDlw']/text()").get()
        #     'author':response.xpath("//a[@class='vGXaw uoMSP kXLw7 R6ToQ JVs7s R6ToQ']/text()").get()
        #     'published':response.xpath("//time/@datetime").get()
        #     'url':response.url
        #     'photos':response.xpath("//img[@class='I7OuT DVW3V L1BOa']/@srcset").getall()
        # }
        loader=ItemLoader(item=SimpleItem(),response=response)
        loader.add_xpath('title',"//p[@class='liDlw']/text()")
        loader.add_xpath('author',"//a[@class='vGXaw uoMSP kXLw7 R6ToQ JVs7s R6ToQ']/text()")
        loader.add_xpath('published',"//time/@datetime")
        loader.add_value('url',response.url)
        loader.add_xpath('photos',"//img[@class='I7OuT DVW3V L1BOa']/@srcset")
        #Сюда можно было бы еще втулить и категорию, но она будет для всех одинакова - не интересно - втулил автора и дату публикации!
        
        yield loader.load_item()
        # yield item
    
    # Закрытие соединения с MongoDB после завершения работы паука P.S. - это с предыдущего семинара. В этот раз не складывал в БД, а можно было бы!
    def closed(self, reason):
        print("Done")
        # self.client.close()
```

```python
#Запускающий модуль - также все по инструкции из семинара
from scrapy.crawler import CrawlerProcess
from scrapy.utils.reactor import install_reactor
from scrapy.utils.log import configure_logging
from scrapy.utils.project import get_project_settings
from myspider import UnsplashSpider
from pipelines import SimplePipeline,SimplePhotosPipeline

configure_logging()
install_reactor('twisted.internet.asyncioreactor.AsyncioSelectorReactor')
process=CrawlerProcess(settings={
    'USER_AGENT': 'My Custom Spider User Agent',
    'DOWNLOAD_DELAY': 0.2,# не щадим сайты!
    'FEEDS': {'info.csv': {'format': 'csv'}},
    'IMAGES_STORE': 'photos',  # Папка для сохранения изображений
    'ITEM_PIPELINES': {
        '__main__.SimplePipeline': 300,  # Указываем pipeline и его приоритет
        '__main__.SimplePhotosPipeline':200, #чем меньше тем первее 
    }
})
category=input("Введите категорию") #Запрос на ввод категории фоток
process.crawl(UnsplashSpider,category=category)
process.start()
```
